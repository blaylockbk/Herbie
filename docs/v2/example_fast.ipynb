# FastHerbie - Build-your-own-grib files

FastHerbie is a major change since its original implementation. It locates many grib files, iterating over many model initialization times and/or forecast lead times (step). For the successfully resolved file locations, you may subset those files for specific fields. 

- Use `FH.inventory` to list a concatted DataFrame of fields from all files.
- Use `FH.download` 

```python
from herbie.v2 import FastHerbie, HRRR
from datetime import datetime, timedelta
```

```python
# Create a FastHerbie object, iterating over model initialization dates
# and forecast lead times (steps)
FH = FastHerbie(
    [
        datetime(2026, 2, 1, 1),
        datetime(2026, 2, 1, 2),
        datetime(2026, 2, 1, 3),
    ],
    step=range(0, 3),
    model=HRRR,
)
FH
```

```python
FH.inventory("TMP:2 m")
```

```python
file = FH.download("TMP:2 m", overwrite=True)
file
```

```python
# FastHerbie makes an index file
# If wgrib2 exists, makes a wgrib2-style file, else, makes an eccodes-style file
file_idx = file.with_name(file.name + ".idx")

print(file)
print(file_idx)
print("----Generated Index File-------------------------------")
print(file_idx.read_text())
```

## Partitioned files

```python
file = FH.download("TMP:2 m", partition_by="step")
file
```
