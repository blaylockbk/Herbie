```python
H = Herbie("2025-11-01")
H.idx
```

```python
%%time
df = (
    pl.read_csv(
        H.idx,
        has_header=False,
        separator=":",
        new_columns=[
            "grib_message",
            "start_byte",
            "reference_time",
            "variable",
            "level",
            "forecast_time",
        ],
    )
    .with_columns(
        pl.col("reference_time").str.pad_end(14, "0").str.to_datetime("d=%Y%m%d%H%M")
    )
    .insert_column(2, (pl.col("start_byte").shift(-1) - 1).alias("end_byte"))
    .insert_column(
        3,
        pl.concat_str(
            pl.col("start_byte"),
            pl.lit("-"),
            pl.col("end_byte").cast(pl.String).fill_null(""),
        ).alias("byte_range"),
    )
    .insert_column(0, pl.lit(H.idx).alias("url"))
)

# Drop column with all nulls
df = df[[s.name for s in df if not (s.null_count() == df.height)]]
df
```

```python
df1 = df.filter(pl.col("variable").is_in(["UGRD", "VGRD"]))
df1
```

```python
df2 = (
    df1.with_columns(
        download_group=(pl.col("grib_message").diff().fill_null(1) != 1).cum_sum(),
    )
    .group_by("url", "download_group", maintain_order=True)
    .agg(
        pl.col("start_byte").min(),
        pl.col("end_byte").max(),
        pl.col("grib_message"),
        pl.col("variable"),
        pl.col("level"),
    )
    .insert_column(
        3,
        pl.concat_str(
            pl.col("start_byte"),
            pl.lit("-"),
            pl.col("end_byte").cast(pl.String).fill_null(""),
        ).alias("byte_range"),
    )
)
df2
```
