```python
from rich.console import Console
from rich.text import Text
from datetime import datetime

model = "hrrr"
product = "prs"
date = datetime(2025, 12, 1)
fxx = 4
grib_source = "aws"

# Create rich text with styling
text = Text()

# Herbie logo
text.append("▌", style="bold red on white")
text.append("▌", style="bold blue on #f0ead2")
text.append("Herbie", style="bold black on #f0ead2")
text.append(" ")

# Model name
text.append(f"{model.upper()} model ", style="white")

# Product
text.append(f"{product}", style="italic")
text.append(" product initialized ", style="white")

# Date and forecast hour
text.append(f"{date:%Y-%b-%d %H:%M UTC}", style="green")
text.append(f" F{fxx:02d}", style="bright_green")

# Source
text.append(" ┊ ", style="white")
text.append(f"source={grib_source}", style="italic #ff9900")

text
```
