# Open hypercube in a DataTree

For datasets with multiple hypercubes (different vertical levels), cfgrib returns a list of datasets, but Herbie returns an xarray DataTree with nodes named with the vertical level type.

```python
from herbie.v2 import Herbie
from datetime import datetime, timedelta
```

```python
H = Herbie.HRRR(datetime(2026, 1, 1))
H
```

```python
H.inventory("(TMP:2 m|GRD:10 m)")
```

```python
ds = H.xarray("(TMP:2 m|GRD:10 m)")
ds
```

```python
ds["/heightAboveGround_10"]
```

```python
H.inventory("UGRD:.*0 ")
```

```python
H.xarray("UGRD:.*0 ")
```
